## 1389. Create Target Array in the Given Order

### Example 1

```python
nums = [0,1,2,3,4]
index = [0,1,2,2,1]
```

```python
arr = []
for i, v in zip(index, nums):
    arr.insert(i, v)
    print(arr)
```
