## 1. Two Sum

### Example 1

```python
# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
```

```python
nums = [2, 7, 11, 15]
target = 9
```

```python
hashmap = {}

for i, value in enumerate(nums):
    #print(i, value)
    remaining = target - nums[i]
    #print(remaining)

    if remaining in hashmap:
        print(i, hashmap[remaining])
    else:
        hashmap[value] = i
        #print(hashmap)
```
