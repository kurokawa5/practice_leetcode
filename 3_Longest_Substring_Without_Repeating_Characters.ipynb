## 3. Longest Substring Without Repeating Characters

### Example 1

```python
s = "abcabcbb"
```

```python
chars = {}
start = 0
max_length = 0
            
for end, c in enumerate(s):
  print(end, c)
  
  if c in chars:
    start = max(start, chars[c] + 1)
            
  max_length = max(max_length, end - start + 1)
  
  chars[c] = end
  
print(max_length)
```
