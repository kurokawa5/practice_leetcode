## 20. Valid Parentheses

### Example 1

```python
s = "()"
```

```python
d = {'(':')', '{':'}','[':']'}
stack = []

for i in s:
    print("i", i)
    if i in d:  # 1
        stack.append(i)

    elif len(stack) == 0 or d[stack.pop()] != i:  # 2
        print(False)

    print("stack", stack)
    print("len(stack)", len(stack))

print(len(stack) == 0) # 3
```
