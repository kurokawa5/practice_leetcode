## 118. Pascal's Triangle

### Example 1

```python
numRows = 5
```

```python
if numRows == 1: 
    print([[1]])

tri = [[1]]

for i in range(1, numRows):
    #print("i", i)
    row = [1]

    for j in range(1, i):
        #print("j", j)
        row.append(tri[i-1][j-1] + tri[i-1][j])
        #print(tri[i-1][j-1])
        #print(tri[i-1][j])
        #print("tri[i-1][j-1] + tri[i-1][j]", tri[i-1][j-1] + tri[i-1][j]) 
    
    row.append(1)
    #print(row)
    tri.append(row)

print(tri)
```
